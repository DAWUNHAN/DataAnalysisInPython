# 서브셋 가져오기

## 서울 데이터만 가져오기

```python
# 시도명이 서울특별시인 경우, df_seoul에 넣어준다.

df_seoul = df[df["시도명"] == "서울특별시"].copy()
df_seoul.shape
```

## 특정 텍스트가 들어가는 데이터만 가져오기

```python
# "상호명_소문자" 칼럼을 생성

df_seoul["상호명_소문자"] = df_seoul["상호명"].str.lower()
```

```python
# 상호명_소문자가 "이디야|이디아|ediya" 인 경우

df_seoul.loc[df_seoul["상호명_소문자"].str.contains("이디야|이디아|ediya"), "상호명_소문자"].shape
```

```python
# 상호명_소문자가 "스타벅스|starbucks"인 경우

df_seoul.loc[df_seoul["상호명_소문자"].str.contains("스타벅스|starbucks"), "상호명_소문자"].shape
```

```python
# 상호명_소문자가 "이디야|이디아|ediya|스타벅스|starbucks" 인 경우, df_cafe에 넣어주기.

df_cafe = df_seoul[df_seoul["상호명_소문자"].str.contains("이디야|이디아|ediya|스타벅스|starbucks")].copy()
df_cafe.shape
```

```python
# 상호명_소문자가 스타벅스인 경우 브랜드명을 스탁벅스로 넣는다.

df_cafe.loc[df_cafe["상호명_소문자"].str.contains("스타벅스|starbucks"), "브랜드명"] = "스타벅스"
df_cafe[["상호명_소문자", "브랜드명"]].head()
```

```python
# 상호명_소문자가 스타벅스가 아닌 경우, 이디야로 넣어주기.

df_cafe.loc[~df_cafe["상호명_소문자"].str.contains("스타벅스|starbucks"), "브랜드명"] = "이디야"
df_cafe[["상호명_소문자", "브랜드명"]].head()
```

```python
# tail로 미리보기

df_cafe[["상호명", "브랜드명"]].tail()
```

# 구별 브랜드별 매장 수

## groupby 로 구별 매장 개수 찾기

```python
# 시군구명에 따라 브랜드명의 개수를 구한다
# groupby는 시리즈 형태로 반환되므로 dataframe으로 만들어야 한다.

df_cafe_vs = df_cafe.groupby(["시군구명", "브랜드명"])["상호명"].count()
df_cafe_vs.head()
```

```python
# 데이터 프레임으로 만들기 (to_frame())
# 멀티인덱스 형태로 나온다.

df_cafe_vs.to_frame().head()
```

```python
# 데이터 프레임으로 만들기 (reset_index())

df_cafe_vs = df_cafe_vs.reset_index()
df_cafe_vs.head()
```

```python
df_cafe_vs.columns = ["구", "브랜드명", "매장수"]
df_cafe_vs.head()
```

## pivot_table로 구별 매장 수 구하기

```python
# pivot 은 연산 불가, pivot_table은 연산 가능

pivot = pd.pivot_table(data=df_cafe, index =["시군구명", "브랜드명"], values="상호명", aggfunc='count')
pivot.head()
```

# folium으로 지도 시각화하기

```python
import folium
```

```python
# 지도의 중심 정하기

lat = df_cafe["위도"].mean()
long = df_cafe["경도"].mean()
lat, long
```

## 스타벅스, 이디야 카페 매장 표시

```python
m = folium.Map([lat, long], zoom_start=12, tiles="Stamen Toner")

for i in df_cafe.index:
    sub_lat = df_cafe.loc[i, "위도"]
    sub_long = df_cafe.loc[i, "경도"]
    
    title = f"{df_cafe.loc[i, '상호명']}  -  {df_cafe.loc[i, '도로명주소']}"
    
    color = "green"
    
    if df_cafe.loc[i, "브랜드명"] == "이디야":
        color = "blue"
    
    folium.CircleMarker([sub_lat, sub_long],
                        radius = 3, 
                        color = color,
                         tooltip=title).add_to(m)
m 
```

## choropleth를 구현하기

- 행정 구역 별로 나누기 위해서 JSON파일을 로드한다

```python
geo_path = 'data/seoul_municipalities_geo_simple.json'

import json
geo_json = json.load(open(geo_path, encoding="utf-8"))
geo_json["features"][0]['properties']['name']
```

```python
# geo_json의 구 명칭과 df_cafe_vs의 구 명칭을 매칭 시킨다.

df_star = df_cafe_vs[df_cafe_vs["브랜드명"] == "스타벅스"].copy()
df_star.head()
```

```python
# geo_json에서 구 이름 가져오기

geo_json["features"][0]["properties"]["name"]
```

## 스타벅스 매장 지도에 표현

```python
m = folium.Map([lat, long], zoom_start = 11)

folium.Choropleth(
    geo_data = geo_json,
    name = 'Choropleth',
    data = df_star,
    columns = ['구', '매장수'],
    key_on = 'feature.properties.name',
    fill_color = 'Greens',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = '스타벅스 매장 수'
).add_to(m)
m
```

## 이디야 매장 지도에 표현

```python
df_ediya = df_cafe_vs[df_cafe_vs["브랜드명"] == "이디야"].copy()
df_ediya.head()
```

```python
m = folium.Map([lat, long], zoom_start = 11)

folium.Choropleth(
    geo_data = geo_json,
    name = 'Choropleth',
    data = df_ediya,
    columns = ['구', '매장수'],
    key_on = 'feature.properties.name',
    fill_color = 'Blues',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = '이디야 매장 수'
).add_to(m)
m
```

## 매장 수를 반영하기

```python
df_vs = df_cafe_vs.pivot(index = "구", columns="브랜드명", values="매장수")
df_vs["매장수비교"] = df_vs["스타벅스"] - df_vs["이디야"] > 0
df_vs.head()
```

```python
# choropleth 그릴 때, radius type이 float이므로, float으로 만든다. 

df_vs["매장수비교"] = df_vs["매장수비교"].astype(float)
df_vs.head()
```

```python
df_vs.index
```

```python
df_vs = df_vs.reset_index()
df_vs.head()
```

## Choropleth로 표현하기

```python
# 스타벅스가 더 많은 구가 진한색, 이디야가 많은 곳은 연한색.

m = folium.Map([lat, long], zoom_start = 11)

folium.Choropleth(
    geo_data = geo_json,
    name = 'Choropleth',
    data = df_vs,
    columns = ['구', '매장수비교'],
    key_on = 'feature.properties.name',
    fill_color = 'BuGn',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = '매장 수 비교'
).add_to(m)


m
```

```python
# 위경도의 평균 값 구하기
    
df_gu_mean = df_cafe.pivot_table(index="시군구명", values=["경도", "위도"], aggfunc="mean")
df_gu_mean.reset_index()
df_gu_mean.head()
```

```python
df_vs = df_vs.merge(df_gu_mean, how='inner', left_on='구', right_on='시군구명')
df_vs.head()
```

```python
for i in df_vs.index:
    sub_long = df_vs.loc[i, "경도"]
    sub_lat = df_vs.loc[i, "위도"]
    
    print(df_vs.loc[i, "구"], sub_long, sub_lat)
```

```python
m = folium.Map([lat, long], zoom_start = 11, tiles='Stamen toner')

folium.Choropleth(
    geo_data = geo_json,
    name = 'Choropleth',
    data = df_vs,
    columns = ['구', '매장수비교'],
    key_on = 'feature.properties.name',
    fill_color = 'BuGn',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = '매장 수 비교'
).add_to(m)

for i in df_vs.index:
    sub_long = df_vs.loc[i, "경도"]
    sub_lat = df_vs.loc[i, "위도"]
    
    cafes = ["스타벅스", "이디야"]
    for cafe in cafes:
        cafe_count = df_vs.loc[i, cafe]
        
        gu = df_vs.loc[i, "구"]
        tooltip = f"{gu} {cafe} : {cafe_count}"
        radius = np.sqrt(cafe_count) * 3
        
        color = "green"
        if cafe == "이디야":
            sub_long = sub_long + 0.01
            color = "blue"
            
        folium.CircleMarker([sub_lat, sub_long],
                            radius = radius,
                            color = color,
                            fill = True,
                            tooltip = tooltip
                           ).add_to(m)

m
```

```python
m.save("스타벅스와 이디야 매장 개수 비교.html")
```
