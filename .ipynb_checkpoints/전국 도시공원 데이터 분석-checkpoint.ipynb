## 편익시설

```python
df['공원보유시설(편익시설)'].value_counts()
```

```python
# 토큰화를 위해 각 시설을 ','로 구분하는 함수 구현

def get_string(text):
    if pd.isna(text):
        return text
    
    text = re.sub("[\d\.'\-(nan)\[\]\/]", "", text)
    text = re.sub(" ", "", text)
    text = re.sub(",{2,}", ",", text)
    return text
```

```python
get_string("[1. '화장실' 5, 2. 주차장, -, nan, /]")
```

```python
get_string(np.nan)
```

```python
df["편익시설"] = df['공원보유시설(편익시설)'].map(get_string)
df["편익시설"].head()
```

```python
df["편익시설"].value_counts().tail()
```

```python
type(np.nan)
```

```python
# nan의 타입이 float이므로 바로 join으로 연결 불가. -> notnull만 가져온다

fac = ",".join(df.loc[df["편익시설"].notnull(), "편익시설"])
fac = get_string(fac)
fac[:100]
```

```python
# 위의 내용을 다른 방법으로 해보기. 
# 리스트 형태로 만들고 str로 변환

fac = str(df["편익시설"].tolist())
fac = get_string(fac)
fac[:100]
```

# 특정 정보 마스킹

## 전화번호 마스킹

```python
# 가운데가 4자리일 때 번호 마스킹 "010-1234-5678"
# re.sub("규칙", "패턴", "데이터")
# 규칙
# (\d{3}) 는 숫자 3자리를 의미.
# (\d{1,2}) 는 숫자 1~2자리를 의미.
# 패턴
# r"\1-\2**-**\5" 의 "\숫자"는 앞에서 () 안에 정의한 값의 순서를 의미. 
# 변수의 순서라고 생각해도 된다.
# 여기에서 \3, \4는 쓰지 않고 * 로 대체.
# r은 raw를 의미한다.

re.sub("(\d{2})-(\d{2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5", "010-1234-5678")
```

```python
# 가운데 번호가 세 자리 일 때

re.sub("(\d{2})-(\d{1,2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5", "010-123-4567")
```

```python
# 가운데 번호가 2자리 일 때

re.sub("(\d{2})-(\d{1, 2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5", "010-12-3456")
```

```python
# 함수를 만들어서 구현한다

phone = "+82-10-1234-5678"

def mask_phone_number(phone):
    
    if pd.isnull(phone):
        return phone
    
    num = phone.split("-")
    num_len = len(num) // 2
    
    num[num_len] = re.sub("\w", "*", num[num_len])
    num[-1] = re.sub("(\d{2})(\d{2})", r"**\2", num[-1])
    
    return "-".join(num)

mask_phone_number(phone)
```

```python
# 결측치가 들어간 경우 확인

mask_phone_number(np.nan)
```

```python
# map을 통해 함수 적용

df["전화번호(전처리)"] = df["전화번호"].map(mask_phone_number)
df[["공원명", "전화번호", "전화번호(전처리)"]] 
```

## 이메일 주소 마스킹

```python
# 전부 다 마스킹

email = "address@mail.com"

re.sub("[^@.]", "x", email)
```

```python
# 아이디만 마스크 처리

email = "address@mail.com"

address, domain = email.split('@')
mask = re.sub("\w", "*", address[1:-1])
print(f"{address[0]}{mask}{address[-1]}@{domain}")
print(email)
```

```python
def mask_email(email):
    address, domain = email.split("@")
    mask = re.sub("\w", "*", address[1:-1])
    return f"{address[0]}{mask}{address[-1]}@{domain}"

mask_email(email)
```

```python
mask_email("1234address_123@gmail.com")
```
